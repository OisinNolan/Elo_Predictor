# KNN regression for predicting Elo

### Choosing value for k

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt
from sklearn.dummy import DummyRegressor

X = np.load('../data/knn_games_1k_10.npy')
y = np.load('../data/knn_y_1k.npy')

K = range(1, 100, 5)
accuracy = []
std = []

for k in K:
    model = KNeighborsRegressor(n_neighbors = k, weights = 'uniform', metric='hamming')
    scores = cross_val_score(model, X, y, cv=5, scoring='r2')
    accuracy.append(scores.mean())
    std.append(scores.std())
    
plt.errorbar(K, accuracy, yerr=std)
plt.xlabel('k')
plt.ylabel('r2')
plt.gcf().set_size_inches(10, 6)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_predict

X = np.load('../data/knn_games_1k_10.npy')
y = np.load('../data/knn_y_1k.npy')


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

K = range(8, 20, 2)
mse = []
r2 = []

for k in K:
    model = KNeighborsRegressor(n_neighbors=k, weights='uniform', metric='hamming')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse.append(mean_squared_error(y_test, y_pred))
    
plt.errorbar(K, mse)
plt.xlabel('k')
plt.ylabel('mse')
plt.gcf().set_size_inches(10, 6)
plt.show()
```

# Model Comparison
- 1000 training data points
- Training data here contains only opening (first 10 moves)
- Hamming distance appears to consistently perform better. This is expected as data is catagorical

```python
from sklearn.linear_model import LinearRegression

X = np.load('../data/knn_games_1k_10.npy')
y = np.load('../data/knn_y_1k.npy')

knn_hamming_mse = []
knn_hamming_r2 = []
dummy_mse = []
dummy_r2 = []

for _ in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    knn_hamming = KNeighborsRegressor(n_neighbors=12, weights='uniform', metric='hamming')
    knn_hamming.fit(X_train, y_train)
    hamming_y_pred = knn_hamming.predict(X_test)
    knn_hamming_mse.append(mean_squared_error(y_test, hamming_y_pred))
    knn_hamming_r2.append(r2_score(y_test, hamming_y_pred))

    baseline = DummyRegressor()
    baseline.fit(X_train, y_train)
    baseline_y_pred = baseline.predict(X_test)
    dummy_mse.append(mean_squared_error(y_test, baseline_y_pred))
    dummy_r2.append(r2_score(y_test, baseline_y_pred))

print(f'MSE KNN Hamming: {np.mean(knn_hamming_mse)}')
print(f'R2 KNN Hamming: {np.mean(knn_hamming_r2)}')
print(f'MSE Dummy: {np.mean(dummy_mse)}')
print(f'R2 Dummy: {np.mean(dummy_r2)}')
```
